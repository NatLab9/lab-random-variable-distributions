```python
import pandas as pd
```

```python
data =  pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
#1. Get the numerical variables from our dataset.
numerical = data._get_numeric_data()
numerical.head()
```

```python
#2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
import matplotlib.pyplot as plt
def plot_continuous_variables(numerical):
    
    for col in numerical.columns:
        plt.figure()
        plt.title(col)
        numerical[col].plot(kind='hist')
        
plot_continuous_variables(numerical)
```

```python
#All fof the distribitions are skewed, exponentially decreasing.
#Only Months Since Policy Inception might seem almost normal 
```

```python
import numpy as np

# apply log-transform to numerical columns
import pandas as pd
import numpy as np

# replace non-positive values with a small positive value
numerical[["Income", "Monthly Premium Auto", "Months Since Last Claim",
      "Months Since Policy Inception", "Number of Open Complaints",
      "Number of Policies", "Total Claim Amount"]] = \
    numerical[["Income", "Monthly Premium Auto", "Months Since Last Claim",
          "Months Since Policy Inception", "Number of Open Complaints",
          "Number of Policies", "Total Claim Amount"]].apply(lambda x: np.where(x <= 0, 1e-6, x))

# apply log-transform to numerical columns
numerical[["Income", "Monthly Premium Auto", "Months Since Last Claim",
      "Months Since Policy Inception", "Number of Open Complaints",
      "Number of Policies", "Total Claim Amount"]] = \
    np.log(numerical[["Income", "Monthly Premium Auto", "Months Since Last Claim",
                 "Months Since Policy Inception", "Number of Open Complaints",
                 "Number of Policies", "Total Claim Amount"]])

numerical.head()
```

```python
import matplotlib.pyplot as plt
def plot_continuous_variables(numerical):
    
    for col in numerical.columns:
        plt.figure()
        plt.title(col)
        numerical[col].plot(kind='hist')
        
plot_continuous_variables(numerical)
```
